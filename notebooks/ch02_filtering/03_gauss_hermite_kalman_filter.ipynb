```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from filters.kalman_filter import *
from filters.extended_kalman_filter import *
from filters.gauss_hermite_kalman_filter import *
from models.pendulum import *
```

## Define the Pendulum

```python
dt = 0.01
R = np.array([[0.1]]) # measurement noise

pendulum_system = make_true_pendulum_system(dt = dt,
                                            R = R)
```

## Generate Data

```python
# ============================================================
# 0. Random number generator
# ============================================================
rng = np.random.default_rng(seed=None)  # non-deterministic runs

# ============================================================
# 1. Ground-truth system definition (reality)
# ============================================================
x0 = np.array([1.5, 0.0])  # true initial state

T = 5.0
Nsims = int(T / dt)
times = np.linspace(0.0, T, Nsims + 1)

# Observe at every step except t0
delta = 1
obs_ind = np.arange(delta,times.shape[0], delta)

# ============================================================
# 2. Generate ground truth and synthetic sensor data
# ============================================================
truth, measurements = generate_nonlinear_data(
    times=times,
    x0=x0,
    obs_ind=obs_ind,
    system=pendulum_system,     # TrueNonLinearSystem
    rng=rng
)

plot_data_and_truth(
    truth,
    measurements,
    title="Ground Truth and Sensor Measurements"
)
```

## Gauss-Hermite Kalman Filter ( Order 3)

```python
# ============================================================
# 3. Gauss Hermite Kalman filter (Order 3 )design (estimator's belief)
# ============================================================
# Prior belief
X0 = Gaussian(
    mean=x0,
    cov= np.eye(2)
)

dt = 0.01
qc = 0.1

ghkf_model = GaussHermiteKalmanFilterModel(
    Phi = lambda x : pendulum_dynamics(x, dt),
    h = pendulum_measurement,
    Q = pendulum_process_noise_cov(dt, qc),
    R = R,
    num_pts=3
)

# ============================================================
# 4. Run Gauss-Hermite (Order 3) Kalman filter and visualize estimation
# ============================================================
kf_history = run_gauss_hermite_kalman_filter(
    y_Observations=measurements,
    model=ghkf_model,
    X0=X0
)

plot_data_and_truth(
    truth,
    measurements,
    kf=kf_history,
    title="Gauss-Hermite (Order 3) Kalman Filter State Estimation"
)
```

```python
# ============================================================
# 3. Gauss Hermite Kalman filter (Order 5 )design (estimator's belief)
# ============================================================


ghkf_model_order5 = GaussHermiteKalmanFilterModel(
    Phi = lambda x : pendulum_dynamics(x, dt),
    h = pendulum_measurement,
    Q = pendulum_process_noise_cov(dt, qc),
    R = R,
    num_pts=5
)

# ============================================================
# 4. Run Gauss-Hermite (Order 5) Kalman filter and visualize estimation
# ============================================================
kf_history_order5 = run_gauss_hermite_kalman_filter(
    y_Observations=measurements,
    model=ghkf_model_order5,
    X0=X0
)

plot_data_and_truth(
    truth,
    measurements,
    kf=kf_history_order5,
    title="Gauss-Hermite (Order 5) Kalman Filter State Estimation"
)
```
